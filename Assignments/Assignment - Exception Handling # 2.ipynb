### Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

Exception class is the base class for all Exceptions in Python, whether it is built-in or Custom Exceptions. Exception class is inherited by the user-defined classes to handle raising an exception.
Any custom exception that you create, will always be a sub-class of Exception class.


### Q2. Write a python program to print Python Exception Hierarchy.

BaseException class is the parent class of all Exception classes in Python. We can recursively go through __subclasses__() down by the inhertitance tree.

```python
def class_tree(cls, indent=0):
    
    print('-'*indent, cls.__name__)
    
    for subcls in cls.__subclasses__():
        class_tree(subcls, indent + 3)

class_tree(BaseException)
```

### Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

This exception type is raised for arithmetic related errors. This base class has the exceptions OverflowError, FloatingPointError, ZeroDivisionError.

**OverflowError:** An OverflowError Exception is raised when an arithmetic operation exceeds the limit to be represented.

**FloatingPointError:** This exception occurs when a floating-point calculation is attempted that is not supported by the available hardware or software. It occurs when trying to calculate a number that is too large or too small to represent as a floating-point number. Python does not raise FloatingPointError error by default. We can use nump.errstate to get FloatingPointError in the output.

```python
# this will raise OverflowError Exception
j = 5.0

for i in range(1, 1000):
    j = j**i #the exponentiation opertor
    print(j)
```

```python
import numpy as np
with np.errstate(invalid='raise'):
    print(np.sqrt(-1))
```

### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError is the base class for exceptions that are raised when a key or index used on a mapping or sequence is invalid. It has two sub-classes: IndexError and KeyError.

**IndexError:** Raised when a sequence subscription is out of range.

**KeyError:** Raised when you try to access a key that isn't in a dictionary.

```python
a = [11,22,33,44,66]

print(a[6]) # this will raise IndexError Exception
```

```python
ages = {'Jim': 30, 'Pam': 34, 'Kevin': 19}
print(ages['Mike']) # this will raise a KeyError Exception
```

### Q6. List down some best practices for exception handling in python.

 - **Catching Specific Exceptions instead of Generic ones:** This is crucial for writing maintainable and easy-to-debug code.
 - **Logging Exceptions:** Logging plays a crucial role in capturing valuable information about the error and the context in which it occurred. This can greatly assist in diagnosing and fixing issues in your code.
 - **Be descriptive in Error Handling:** Use meaningful error messages and log exception details. This will make it easier to understand and fix the problem.
 - **Avoid empty except block:** Avoid using an empty except block, because it will catch all exceptions, including those that you didn't expect, and make it hard to debug your code.
 - **Use the finally block:** The finally block is used to specify the code that will be executed regardless of whether an exception was raised or not. This can be useful for cleaning up resources, such as closing files or releasing memory.
